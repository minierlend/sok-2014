```python
import pandas as pd
import matplotlib.pyplot as plt
column_names1 = ['Year', 'Petroleum']

df = pd.read_csv('karbon.csv', names=column_names1, skiprows=1)

df.head()
```

```python

df['Year'] = pd.to_datetime(df['Year'], format='%Y')
filtered_df = df[(df['Year'] >= '2023') & (df['Year'] <= '2070')]

# grafen
plt.figure(figsize=(10, 7))
plt.plot(filtered_df['Year'], filtered_df['Petroleum'], marker='o', linestyle='-', color='green', markersize = 4)
plt.title('rlig Petroleum - kvote priser fra R-109/21 (2023-2070)')
plt.xlabel('r')
plt.ylabel('Petroleum - kvote priser i kroner fra R-109/21')
plt.grid(True)
plt.show()
```

```python
df1 = pd.read_csv('naturgass.csv')
column_names2 = ['Month', 'Price']

df1 = pd.read_csv('naturgass.csv', names=column_names2, skiprows=1)

df1.head()


```

```python
import seaborn as sns


df1['Month'] = pd.to_datetime(df1['Month'])

df1['Year'] = df1['Month'].dt.year


yearly_avg = df1.groupby('Year')['Price'].mean().reset_index()

sns.set(style="whitegrid")



plt.figure(figsize=(10, 6))
plt.plot(yearly_avg['Year'], yearly_avg['Price'], marker='o', linestyle='-', color='b')
plt.title('Gjennomsnittlig rlig pris p natur gass i dollar')
plt.xlabel('r', fontsize = 12)
plt.ylabel('Gjennomsnittlig pris', fontsize = 12)
plt.grid(True)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
